```python
# importing modules and libraries 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from xgboost import XGBClassifier
```

```python
# Applying train test split
x = df.drop('Class' , axis = 1)
y = df['Class']
```

```python
x_tr, x_te , y_tr, y_te = train_test_split( x, y , test_size = 0.2 , random_state = 42 , stratify = y )
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42 )
x_tr_res , y_tr_res = sm.fit_resample(x_tr , y_tr )
```

```python
# Building a XGBoost model 

from xgboost import XGBClassifier 
model = XGBClassifier (
    n_estimators = 200 ,
    max_depth = 6 ,
    learning_rate = 0.1 ,
    subsample = 0.8 ,
    colsample_bytree = 0.8 ,
    eval_metric = 'logloss'
)
```

```python
# trainig the model 

model.fit(x_tr_res , y_tr_res )
```

```python
# Once out model have trained we are going to predict the values 
y_pred = model.predict(x_te)
```

```python
y_prob = model.predict_proba(x_te)[ : , 1 ]
y_pred = (y_prob > 0.3 ).astype(int)
```

```python
# now we are going to evaluate our model alright 

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
print (confusion_matrix(y_te , y_pred))
```

```python
# Lets make it validate using roc and auc scale 

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
fpr, tpr, _ = roc_curve(y_te, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# feature importance 
from xgboost import plot_importance
plot_importance(model)
plt.show()
```

```python
# Saving the model for future use cases 
import joblib

joblib.dump(model, "fraud_xgboost_model.pkl")
```
